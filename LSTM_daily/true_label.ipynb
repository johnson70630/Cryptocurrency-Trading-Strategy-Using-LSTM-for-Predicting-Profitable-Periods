## Process data

```python
import pandas as pd
```

```python
df = pd.read_csv('data/processed_price_turning_BTCUSDT.csv')
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['entry_date'])

# Extract the date part from the datetime column
df['date'] = df['datetime'].dt.date

# Group by date and sum the profit for each day
daily_profit = df.groupby('date')['PnL'].sum().reset_index()

daily_profit
```

## Create a new dataset for true label

```python
# Create a new column that flags whether the daily profit is greater than 0
daily_profit['label'] = (daily_profit['PnL'] > 0).astype(int)

# Select only the date and profit_flag columns
new_dataset = daily_profit[['date', 'label']]

new_dataset
```

```python
# Save the DataFrame to a CSV file
new_dataset.to_csv('data/true_label.csv', index=False)

print("DataFrame created and saved to 'data/true_label.csv'.")
```
